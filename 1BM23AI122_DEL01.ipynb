```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

def relu(z):
    return np.maximum(0, z)


class Neuron:
    def __init__(self, weights, bias, activation="sigmoid"):
        self.weights = np.array(weights)
        self.bias = bias
        self.activation = activation
        
    def forward(self, inputs):
        z = np.dot(inputs, self.weights) + self.bias
        if self.activation == "sigmoid":
            return sigmoid(z)
        elif self.activation == "tanh":
            return tanh(z)
        elif self.activation == "relu":
            return relu(z)
        else:
            raise ValueError("Unknown activation function")

inputs = np.array([0.5, -1.2, 3.0])
weights = [0.4, -0.6, 0.3]
bias = 0.1


activations = ["sigmoid", "tanh", "relu"]
outputs = []
for act in activations:
    neuron = Neuron(weights, bias, activation=act)
    outputs.append(neuron.forward(inputs))
    print(f"Activation = {act:7s} --> Output = {outputs[-1]:.4f}")


z = np.linspace(-5, 5, 200)
plt.figure(figsize=(8,6))
plt.plot(z, tanh(z), label="Tanh", color="red", linewidth=2)


plt.title("Activation Functions", fontsize=14)
plt.xlabel("z", fontsize=12)
plt.ylabel("Activation(z)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

z = np.linspace(-5, 5, 200)
plt.figure(figsize=(8,6))
plt.plot(z, sigmoid(z), label="Sigmoid", color="blue", linewidth=2)


plt.title("Activation Functions", fontsize=14)
plt.xlabel("z", fontsize=12)
plt.ylabel("Activation(z)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

z = np.linspace(-5, 5, 200)
plt.figure(figsize=(8,6))
plt.plot(z, relu(z), label="ReLU", color="green", linewidth=2)

plt.title("Activation Functions", fontsize=14)
plt.xlabel("z", fontsize=12)
plt.ylabel("Activation(z)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
