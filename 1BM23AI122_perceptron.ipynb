```python
import math
import matplotlib.pyplot as plt
import numpy as np

class Perceptron:
    def __init__(self, weight=1.0, bias=0.0, activation='linear'):
        self.weight = weight
        self.bias = bias
        self.activation = activation.lower()

    def activate(self, x):
        if self.activation == 'linear':
            return x
        elif self.activation == 'sigmoid':
            return 1 / (1 + np.exp(-x))
        elif self.activation == 'tanh':
            return np.tanh(x)
        else:
            raise ValueError("Unsupported activation function. Choose 'linear', 'sigmoid', or 'tanh'.")

    def predict(self, input_values):
        z = self.weight * input_values + self.bias
        return self.activate(z)

# Function to plot the output of the perceptron for a range of inputs
def plot_perceptron_outputs(weight=1.0, bias=0.0):
    input_range = np.linspace(-10, 10, 500)

    activations = ['linear', 'sigmoid', 'tanh']
    plt.figure(figsize=(10, 6))

    for activation in activations:
        p = Perceptron(weight=weight, bias=bias, activation=activation)
        output = p.predict(input_range)
        plt.plot(input_range, output, label=f'{activation} activation')

    plt.title("Perceptron Output with Different Activation Functions")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run the plot function
if __name__ == "__main__":
    plot_perceptron_outputs(weight=1.0, bias=0.0)
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

class Perceptron:
    def __init__(self, weight=1.0, bias=0.0, activation='linear'):
        self.weight = weight
        self.bias = bias
        self.activation = activation.lower()

    def activate(self, x):
        if self.activation == 'linear':
            return x
        elif self.activation == 'sigmoid':
            return 1 / (1 + np.exp(-x))
        elif self.activation == 'tanh':
            return np.tanh(x)
        elif self.activation == 'relu':
            return np.maximum(0, x)
        elif self.activation == 'softmax':
            # Softmax over the entire input vector (batch)
            exp_x = np.exp(x - np.max(x))  # stability trick
            return exp_x / exp_x.sum()
        else:
            raise ValueError("Unsupported activation function. Choose 'linear', 'sigmoid', 'tanh', 'relu', or 'softmax'.")

    def predict(self, input_values):
        z = self.weight * input_values + self.bias
        return self.activate(z)

# Plot function extended for ReLU and softmax
def plot_perceptron_outputs(weight=1.0, bias=0.0):
    input_range = np.linspace(-10, 10, 500)

    activations = ['linear', 'sigmoid', 'tanh', 'relu', 'softmax']
    plt.figure(figsize=(12, 8))

    for activation in activations:
        p = Perceptron(weight=weight, bias=bias, activation=activation)
        output = p.predict(input_range)
        
        if activation == 'softmax':
            # Softmax output is a probability distribution over input_range, plot it differently
            plt.plot(input_range, output, label=f'{activation} activation')
        else:
            plt.plot(input_range, output, label=f'{activation} activation')

    plt.title("Perceptron Output with Different Activation Functions")
    plt.xlabel("Input")
    plt.ylabel("Output")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run the plot function
if __name__ == "__main__":
    plot_perceptron_outputs(weight=1.0, bias=0.0)
```
